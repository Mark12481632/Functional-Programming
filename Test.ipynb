```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set seed
np.random.seed(1432)
```

```python
# Create data points
theta = np.array([1.23, 0.75])
num_pts = 100

X = np.random.uniform(low=0, high=100, size=num_pts)
y = X*theta[1] + theta[0] + np.random.normal(loc=0.0, scale=5.7, size=num_pts)
```

```python
# Functions:
def error(theta, X, y):
  ones = np.ones((X.shape[0]))
  new_X = np.column_stack((ones, X))
  diff = new_X @ theta - y
  diff_sq = diff * diff
  return diff_sq.sum()
```

```python
# Plot points
plt.scatter(X, y)
```
